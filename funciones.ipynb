```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5, 5)
e = np.e
def f(x, e):
    return ((e ** x)-(e ** -x))/((e ** x)+(e ** -x))
plt.plot(x, f(x, e), 'r')
plt.xlabel("Valor x")
plt.ylabel("Funcion Sigmoidea(x)")
plt.grid(True)
```

```python
def  escalon(x, c):
    return np.piecewise(x, [x < 0, x > 0], [0, 1])
c = np.linspace(0, 5) 
plt.plot(x, escalon(x,c), 'y')
plt.axis([x[0], x[-1], -0.1, 1.5])
plt.xlabel("Valor x")
plt.ylabel("Funcion Escalon(x)")
plt.grid(True)
```

```python
sigma=0.01
x = np.linspace(-5,5)
e = np.e
pi = np.pi
x = np.linspace(-200, 200)
def f(x):
    return (1/np.sqrt(2*pi*sigma**2))*(e**-((x**2)/2*(sigma**2)))
plt.xlabel("Valor x")
plt.ylabel("Funcion Gaussiana(x)")
plt.plot(x, f(x),':m')
```
